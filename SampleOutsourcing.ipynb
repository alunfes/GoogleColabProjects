```python
import pandas as pd
import numpy as np

# サンプルデータの作成
np.random.seed(42)

# 1か月前のデータ (df_old)
df_old = pd.DataFrame({
    'Engagement ID': range(1, 101),
    'Phase': np.random.choice(['SteadyState', 'Disengagement', 'Onboarding'], 100)
})

# 最新のデータ (df_new)
df_new = df_old.copy()
df_new.loc[df_new.sample(30).index, 'Phase'] = np.random.choice(['SteadyState', 'Disengagement', 'Onboarding'], 30)

# 処理を行うコード
def compare_and_update(df_old, df_new):
    # df_newをコピーして新しいカラムを追加
    result = df_new.copy()
    result['New Status'] = ''

    # 各行を比較して新しいステータスを決定
    for idx, row in result.iterrows():
        old_phase = df_old.loc[idx, 'Phase']
        new_phase = row['Phase']

        if old_phase == new_phase:
            if new_phase == 'SteadyState':
                result.loc[idx, 'New Status'] = 'ss'
            elif new_phase == 'Disengagement':
                result.loc[idx, 'New Status'] = 'dis'
            elif new_phase == 'Onboarding':
                result.loc[idx, 'New Status'] = 'onb'
        else:
            if new_phase == 'SteadyState':
                result.loc[idx, 'New Status'] = 'new ss'
            elif new_phase == 'Disengagement':
                result.loc[idx, 'New Status'] = 'new dis'
            elif new_phase == 'Onboarding':
                result.loc[idx, 'New Status'] = 'new onb'

    return result

# 関数を実行
result_df = compare_and_update(df_old, df_new)

# 結果の表示
print(result_df.head(10))
print("\n状態の集計:")
print(result_df['New Status'].value_counts())
```
